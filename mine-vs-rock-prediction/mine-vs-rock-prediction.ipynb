```python
# import all dependencies
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
sonar = pd.read_csv('sonar-data.csv', header=None)
sonar
```

```python
# find the rows and columns of the dataframe
sonar.shape
```

```python
# check for missing data
sonar.info()
```

```python
# find the aggregate functions for all columns
sonar.describe()
```

```python
# find the amount of rock data
len(sonar[sonar.iloc[:, 60] == 'R'])
```

```python
# find the amount of mine data
len(sonar[sonar.iloc[:, 60] == 'M'])
```

```python
# find the mean from the data
sonar.groupby(60).mean()
```

```python
# find the min from the data
sonar.groupby(60).min()
```

```python
# find the max from the data
sonar.groupby(60).max()
```

```python
# find the std from the data
sonar.groupby(60).std()
```

```python
# separate into 2 dataframes
X = sonar.drop(60, axis=1)
Y = sonar[60]
```

```python
X
```

```python
Y
```

```python
# train and test data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, stratify=Y, random_state=1)
```

```python
# train the model
model = LogisticRegression()
model.fit(X_train, Y_train)
```

```python
# evaluate its accuracy on training data
X_train_result = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_result, Y_train)
training_data_accuracy
```

```python
# evaluate its accuracy on testing data
X_test_result = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_result, Y_test)
testing_data_accuracy
```
